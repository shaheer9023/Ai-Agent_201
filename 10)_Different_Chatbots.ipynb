Memorizing Chats

```python
from google.genai.chats import Chat
prompt=input("enter prompt : ")
chat=client.chats.create(
    model=model,

)
while True:
  if prompt=="exit":
    break
  response=chat.send_message([prompt]) # for simple chat
  # response=chat.send_message([prompt,image]) # for image response
  display(Markdown(response.text))
  prompt=input("enter prompt : ")
```

Configure Model Parameters


```python
from google.genai.types import GenerateContentConfig
prompt=input("enter prompt : ")
response=client.models.generate_content(
    model=model,
    contents=prompt,
    config=GenerateContentConfig(
        temperature=0.5,
        top_p=0.95,
        candidate_count=2,
        max_output_tokens=1024,


    )
)
for candidate in response.candidates:
    display(Markdown(candidate.content.parts[0].text))
```

Video Understanding

```python
from google.colab import files

# File upload karne ka prompt show karein
uploaded = files.upload()

# Uploaded file(s) ka naam print karein
for file_name in uploaded.keys():
  print(f"Uploaded file: {file_name}")
```

```python
import time
def uploadVideo(videoFile):
  video=client.files.upload(file=file_name)
  while(video.state == "processing"):
    print("video is processing please wait")
    time.sleep(10)
    video=client.files.get(name=video.name or "")
    if video.state=="failed":
      raise ValueError(video.state)
  print(f"file uploaded :"+ video.uri or "")
  return video
myvideo=uploadVideo(file_name)
```

```python
from google.genai.types import Content,Part

prompt =input("enter prompt : ")
Video=myvideo
response=client.models.generate_content(
    model=model,
    contents=[
        Content(role="user",
                parts=[
                    Part.from_uri(
                        file_uri=Video.uri or "",
                        mime_type=Video.mime_type or ""
                    ),
                ]),

              prompt,
                ]
)

scenes=response.text
Markdown(scenes)
```

```python
from google.genai.types import Content, Part
from gtts import gTTS  # Import gTTS for text-to-speech
from IPython.display import Audio  # For playing the audio in Colab

# Step 1: Convert Text to Audio
tts = gTTS(text=scenes, lang='en')  # Agar aap urdu chahte hain toh 'ur' kar dain
audio_path = "generated_audio.mp3"
tts.save(audio_path)  # Save audio file

# Step 2: Play Audio in Colab
Audio(audio_path)
```
